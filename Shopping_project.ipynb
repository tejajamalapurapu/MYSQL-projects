```python
import mysql.connector
```

```python
def get_dbconnection():
    return mysql.connector.connect(
        host='localhost',
        user='root',
        password='teja@124298',
        database='shopping_db'
    )


def register():
    username=input("Enter username:")
    password=input("Enter password to set:")

    db=get_dbconnection()
    mycursor=db.cursor()
    try:
        mycursor.execute(
            "INSERT INTO users(username,password) VALUES(%s,%s)",(username,password))
        db.commit()
        print("Registration successful")
    except:
        print("username already exists")
    finally:
        db.close()
        

def login():
    username=input("Enter username:")
    password=input("Enter password:")

    db=get_dbconnection()
    mycursor=db.cursor()

    mycursor.execute("SELECT  user_id FROM users WHERE username=%s AND password=%s",(username,password))

    user_id=mycursor.fetchone()

    if user_id:
        print("Login successful")
        return user_id[0]
    else:
        print("Invalid details")
        return None


def view_products():
    db=get_dbconnection()
    mycursor=db.cursor()

    mycursor.execute("SELECT *FROM products")
    products=mycursor.fetchall()
    print("\n----products----")
    for p in products:
         print(f"ID:{p[0]} | {p[1]} | ₹{p[2]} | stock:{p[3]}")
    db.close()

def place_order(user_id):
    product_id=int(input("Enter the product id:"))
    quantity=int(input("Enter the quantity:"))

    db=get_dbconnection()
    mycursor=db.cursor()
    
    try:
        db.start_transaction()
        
        mycursor.execute("SELECT stock FROM products WHERE product_id=%s",(product_id,))
        result=mycursor.fetchone()
        if not result:
            print("Invalid product Id")
            return
        stock=result[0]
        
        
        if stock<quantity:
            print("Sorry!Out of stock")
            return
        
        
        mycursor.execute("INSERT INTO orders (user_id) VALUES(%s)",(user_id,))
        order_id=mycursor.lastrowid
        
        mycursor.execute("INSERT INTO order_items(order_id,product_id,quantity) VALUES(%s,%s,%s)",(order_id,product_id,quantity))
        
        mycursor.execute("UPDATE products SET stock=stock-%s WHERE product_id=%s",(quantity,product_id))
        
        db.commit()
        print("Ordered placed successfully,thanks for shopping")
        
    except Exception as e:
        db.rollback()
        print("Order failed",e)
    finally:
        db.close()

def main():
    while True:
        print("\n1.Register\n2.Login\n3.Exit")
        choice=input("Enter the choice:")
        
        if choice=='1':
            register()
        elif choice=='2':
            user_id=login()
            if user_id:
                while True:
                    print("\n1.View products\n2.place order\n3.Exit")
                    ch=input("Enter the choice:")
                    
                    if ch=='1':
                        view_products()
                    elif ch=='2':
                        place_order(user_id)
                    elif ch=='3':
                        print("Logged out")
                        break
                    else:
                        print("Invalid option")
        elif choice=='3':
            break
            


if __name__=="__main__":
    main()
```
